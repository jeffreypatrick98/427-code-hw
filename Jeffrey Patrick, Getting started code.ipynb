```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#histogram of normalized random distribution
x = np.random.normal(5, 10, 100)
plt.hist(x)
plt.show()

#historgram of exponential random distribution
x_exp=np.random.exponential(1, 1000)
y_exp=np.random.exponential(1, 1000)
z_exp=np.random.exponential(1, 1000)
plt.hist(x_exp)
plt.show()

#3D scatter of normalized random distribution
y = np.random.normal(5, 10, 100)
z = np.random.normal(5, 10, 100)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()

#3D scatter of exponential random distribution
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x_exp, y_exp, z_exp, c='b', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```
